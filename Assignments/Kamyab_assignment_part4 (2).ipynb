```python
#!pip install numpy==1.21.4
#!pip install pandas==1.3.4
#!pip install keras==2.1.6
#!pip install scikit-learn
import itertools
import numpy as np
import keras
from keras.models import Sequential
from keras.layers import Dense
import pandas as pd
import sklearn
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import mean_squared_error
```

now we open the data and seperate target and predictors

```python
concrete_data = pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv')
concrete_data_columns = concrete_data.columns
predictors = concrete_data[concrete_data_columns[concrete_data_columns != 'Strength']] # all columns except Strength
target = concrete_data['Strength'] # Strength column
```

But before we fit the data to the model we first normalize our data

```python
predictors_norm = (predictors - predictors.mean()) / predictors.std()
```

we will continue to use the normalized data 
the rest of the steps are similar to previous parts

we need to find out the number of predictor columns as we will use this for the input layer

```python
n_cols = predictors_norm.shape[1] # number of predictors
print('number of inputs:' , n_cols)
```

now we define a function to create a model

the diffrence here is that our network has 3 hidden layers instead of just 1


```python
def init_model():
    model = Sequential()
    model.add(Dense(10, activation = 'relu', input_shape = (n_cols,)))
    model.add(Dense(10, activation = 'relu'))
    model.add(Dense(10, activation = 'relu'))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

now we follow step C using 100 epochs and normalized data

more layers should result in a better model.

Let's see!

```python
msq_list = []
y_pred = []
for _ in itertools.repeat(None,50):
   #step1 split data
    X_train, X_test, y_train, y_test = train_test_split( predictors_norm, target, test_size=0.3, random_state=4)
    #step 2 train model
    model = init_model()
    model.fit(X_train, y_train ,validation_split=0.3, epochs=100, verbose=2)
    #step 3 predict, evaluate and append to list
    y_pred = model.predict(X_test)
    msq_list.append(mean_squared_error(y_test,y_pred))
```

now we calculate the mean and the standard deviation of the mean squared error

```python
mean = np.mean(msq_list)
std = np.std(msq_list)
print('the mean of the MSQ errors was {a} and the standard deviation was {b}'.format(a = mean, b= std))
```

comparing with the previous models, our new model works much better and results in a much smaller error
